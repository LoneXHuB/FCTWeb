```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data as a Pandas DataFrame
df = pd.read_csv('data\sample.txt', sep='\s+')

# Create a new figure with a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot each electrode's data in a separate subplot
for i, ax in enumerate(axs.flatten()):
    col = df.columns[i]
    ax.plot(df[col], lw=2, color=f'C{i}')
    ax.set_title(col, fontsize=14)
    ax.set_xlabel('Time')
    ax.set_ylabel('Amplitude')

    # Set the plot style and background color
    ax.set_facecolor('#f5f5f5')
    ax.grid(color='w', linestyle='-', linewidth=0.5, alpha=0.7)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_linewidth(0.5)
    ax.spines['left'].set_linewidth(0.5)

# Add some padding between the subplots and adjust the overall layout
fig.tight_layout(pad=2)
plt.subplots_adjust(hspace=0.5, wspace=0.3)

# Draw the plot to make sure it's fully rendered
plt.draw()

# Save the plot as a PNG file
plt.savefig('plot.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
